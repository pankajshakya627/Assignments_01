```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from time import time, sleep
import pandas as pd
from bs4 import BeautifulSoup

url = 'https://www.nseindia.com/'
DRIVER_PATH = r"F:\\My_Work02\\Demo\\Web_Scraper_5_min\\chromedriver.exe"

chrome_options = Options()
# chrome_options.add_argument("--headless")  # Comment this line if you want to see the browser window
chrome_options.add_argument("--disable-gpu")  # Needed in headless mode
chrome_options.add_argument("--window-size=1920x1080")  # Set a large window size for more accurate rendering
chrome_options.add_argument("--disable-extensions") #disabling extensions
chrome_options.add_argument("--disable-gpu")  #applicable to windows os only
chrome_options.add_argument("--disable-dev-shm-usage")  # overcome limited resource problems
chrome_options.add_argument("--no-sandbox")  # Bypass OS security model

# Start the ChromeDriver service with the specified path
service = Service(executable_path=DRIVER_PATH)

# Create the WebDriver instance and pass the ChromeOptions
driver = webdriver.Chrome(service=service, options=chrome_options)

driver.get(url=url)
```

```python

ids_list = ["tab1_tableGainer","tab1_tableLoser"]
import random
import time
# Function to introduce random delays between actions
def random_delay(min_delay=1, max_delay=5):
    delay = random.uniform(min_delay, max_delay)
    time.sleep(delay)

# Function to scroll the page
def simulate_scroll(driver):
    # Scroll down
    driver.execute_script("window.scrollBy(0, 300);")
    random_delay(1, 2)


for id_ in ids_list:
    simulate_scroll(driver)
    element = driver.find_element(By.XPATH, f'//div[@id="{id_}"]')

    # Get the HTML content of the div element
    html_content = element.get_attribute('outerHTML')
    html_content
    # Parse the HTML using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract all the tables within the div element
    tables = soup.find_all('table')

    # Convert each table into a DataFrame
    dataframes = []
    for table in tables:
        # Extract table headers
        headers = [th.text.strip() for th in table.find_all('th')]
        # Extract table data
        rows = []
        for row in table.find_all('tr'):
            row_data = [cell.text.strip() for cell in row.find_all(['td', 'th'])]
            rows.append(row_data)
        # Create DataFrame for the table
        df = pd.DataFrame(rows[1:], columns=headers)
        dataframes.append(df)

    # Now 'dataframes' contains a list of DataFrames, each representing a table
    # You can process the DataFrames further or save them as needed
    for idx, df in enumerate(dataframes, start=1):
        print(f"Table {idx}:")
        print(df)
        print("\n")

    # Don't forget to close the driver when you're done
driver.quit()
```
